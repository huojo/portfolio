1 Use gamedata.json for this task. This file contains information of games sold through Steam. Parse out the following information from the data:
- TOP 3 highest metacritic score. Present results using the following format: *Title* has metacritic score of *Score* (for example)
- Games with price discount being 90 % or more. Present results using the following format: *Title* | Discount: *Savings* (for example Metal Gear Solid V: Ground Zeroes | Discount: 90.090090)
- Games having metacritic score higher than steam score. Present results using the following format: *Title* has metacritic score of *MetacriticScore* and steam score of *SteamRatingPercent*

```python
import json
games = []
with open("data_files/gamedata.json") as gamedata:
    games = json.load(gamedata)

```

```python
#TOP 3 highest metacritic score. Present results using the following format: Title has metacritic score of Score (for example)
games.sort(key=lambda x: x["metacriticScore"] , reverse=True)
top3 = games[0:3]
for x in top3:
    title = str(x["title"])
    metascore = int(x["metacriticScore"])
    print(f"{title} has metacritic score of {metascore}")
```

```python
#Games with price discount being 90 % or more. Present results using the following format: Title | Discount: Savings (for example Metal Gear Solid V: Ground Zeroes | Discount: 90.090090)
for g in games:
    title = g["title"]
    savings = float(g["savings"])
    if savings >= 90:
        print(f"{title} | Discount: {savings}")
```

```python
#Games having metacritic score higher than steam score. Present results using the following format: Title has metacritic score of MetacriticScore and steam score of SteamRatingPercent
for g in games:
    title = g["title"]
    metascore = int(g["metacriticScore"])
    steamscore = int(g["steamRatingPercent"])
    if metascore > steamscore:
        print(f"{title} has metacritic score of {metascore} and steam score of {steamscore}")
```

2 Use earthquakes.csv for this task. This file contains information about earthquakes recorded between 1965 and 2016. Earthquake magnitude value describes how strong the earthquake is. Magnitude information can be categorized like presented in the table below (*Source: http://www.geo.mtu.edu/UPSeis/magnitude.html*).

| Magnitude      | Class | Effects |
|----------------|-------|---------|
| 2.4 or less    | Minor | Usually not felt, but can be recorded by seismograph. |
| 2.5 to 5.4     | Light | Often felt, but only causes minor damage. |
| 5.5 to 6.0     | Moderate | Slight damage to buildings and other structures. |
| 6.1 to 6.9     | Strong | May cause a lot of damage in very populated areas. |
| 7.0 to 7.9     | Major | Major earthquake. Serious damage. |
| 8.0 or greater | Great | Great earthquake. Can totally destroy communities near the epicenter. |

Count how many earthquakes have occurred in each class.

<b style="color:red;">Notice:</b> The first value has been modified to be 2.4 or less compared to the original source (has been 2.5 or less).

```python
import pandas as pd

earthquakes = pd.read_csv("data_files/earthquakes.csv",delimiter=",")
earthquakes
```

```python
import numpy as np
conditions = [
    (earthquakes["Magnitude"] <= 2.4),
    (earthquakes["Magnitude"] >= 2.5) & (earthquakes["Magnitude"] <= 5.4),
    (earthquakes["Magnitude"] >= 5.5) & (earthquakes["Magnitude"] <= 6.0),
    (earthquakes["Magnitude"] >= 6.1) & (earthquakes["Magnitude"] <= 6.9),
    (earthquakes["Magnitude"] >= 7.0) & (earthquakes["Magnitude"] <= 7.9),
    (earthquakes["Magnitude"] >= 8.0)
    ]

values = ["Minor", "Light", "Moderate", "Strong","Major","Great" ]

earthquakes["Magnitude Class"] = np.select(conditions,values)

#uusi Magnitude Class on dataframen viimeisenä columnina
earthquakes
```

```python
minor = len(earthquakes[earthquakes["Magnitude Class"] == "Minor"])
light = len(earthquakes[earthquakes["Magnitude Class"] == "Light"])
moderate = len(earthquakes[earthquakes["Magnitude Class"] == "Moderate"])
strong = len(earthquakes[earthquakes["Magnitude Class"] == "Strong"])
major = len(earthquakes[earthquakes["Magnitude Class"] == "Major"])
great = len(earthquakes[earthquakes["Magnitude Class"] == "Great"])

classes = [minor, light, moderate, strong, major, great]
for x in range(len(classes)):
    print(f"The number of {values[x]} class earthquakes is {classes[x]}")
```

3 Use netflix_titles.xml for this task. This file contains information about Netflix movies and TV shows. **Important:** Movies have duration presented in minutes while TV shows have duration presented in amount of seasons! Parse out the following information from the data:
- Movies released in 2017
- TV show and movie amount (present both counts in separate lines)
- Movies with a length between 15 and 20 minutes (values 15 and 20 included)

```python
import xml.etree.ElementTree as e

tree = e.parse("data_files/netflix_titles.xml")
root = tree.getroot()
netflix_collection = []

titles = []
for i in root[0]:
    titles.append(i.tag)

for media in root:
    onetitle = []
    for kama in media:
        onetitle.append(kama.text)
    netflix_collection.append(onetitle)

#into a list of dictionaries
medias = []
for lst in netflix_collection:
    media_obj = {}
    for i in range(len(lst)):
        media_obj[titles[i]] = lst[i]
    medias.append(media_obj)
#medias
```

```python
#Movies released in 2017
movies2017 = []

for m in medias:
    if m["type"] == "Movie" and m["release_year"] == "2017":
        movies2017.append(m["title"])

movies2017
```

```python
#TV show and movie amount (present both counts in separate lines)
movies = []
tvshows = []
for media in medias:
    if media["type"] == "Movie":
        movies.append(media)
    #unsure if there are other types besides movies and tv shows, otherwise I'd use an else here
    if media["type"] == "TV Show":
        tvshows.append(media)

print(f"The number of movies is {len(movies)}")        
print(f"The number of tv shows is {len(tvshows)}")
```

```python
#Movies with a length between 15 and 20 minutes (values 15 and 20 included)
shorts = []
for me in medias:
    if me["type"] == "Movie":
        durationstr = me["duration"]
        duration_and_min = durationstr.split(" ")
        duration = int(duration_and_min[0])
        if duration >= 15 and duration <= 20:
            shorts.append(me["title"])

shorts
```

4 Use the following Rest API for this task: https://tie.digitraffic.fi/api/v1/data/weather-data. Calculate the average for air temperature (ILMA) and humidity (ILMAN_KOSTEUS) values using two decimals.

```python
import requests

endpoint = "/weather-data"
user_id = "5894af975cdc7400113ef7f9"
url = "https://tie.digitraffic.fi/api/v1/data/weather-data" #{0}/{1} #.format(endpoint, user_id)

req = requests.get(url=url)

req.status_code

weather_data = req.json()
#for w in weather_data:
#    print(w)
#ILMAN_KOSTEUS
#ILMA
#weather_data["weatherStations"]
```

```python
airtemps = []
humidities = []
stations =weather_data["weatherStations"][0]["sensorValues"]
for x in range(len(stations)):
    sensor = stations[x]
    sensorname = stations[x]["name"]
    if sensorname == "ILMA":
        airtemps.append(float(sensor["sensorValue"])) 
    if sensorname == "ILMAN_KOSTEUS":
        humidities.append(float(sensor["sensorValue"]))

#calculate the avg for air temperature (ILMA)
avgtemp = sum(airtemps) / len(airtemps)
#calculate avg air humidity (ILMAN_KOSTEUS)
avghum = sum(humidities) / len(humidities)

print(f"Average temperature is {avgtemp}")
print(f"Average humidity is {avghum}")
```
